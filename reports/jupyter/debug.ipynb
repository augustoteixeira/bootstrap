```python
import csv

with open('../../src/single-precision/test.csv', newline='') as csvfile:

    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    md = []
    mf = []
    ln = []
    lf = []
    next(reader, None) # skip header
    for row in reader:
        md.append(float(row[0]))
        mf.append(float(row[1]))
        ln.append(float(row[2]))
        lf.append(float(row[3]))
```

```python
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [17, 10]
plt.plot(lf)
plt.plot(ln, color='red')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [17, 10]
plt.plot(md)
plt.plot(mf, color='red')
plt.show()
```
